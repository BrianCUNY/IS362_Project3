```python
import sqlite3
import pandas as pd
```

```python
# Declaring db file location and connection function
db_path = r"C:\sqlite\db\Chinook_Sqlite.db"
def db_connect(db_path):
    con = sqlite3.connect(db_path)
    return con
```

```python
# Shortened aliases, query to SQL db file, reading data into df, displaying df before manipulations. 
con = db_connect(db_path)
cur = con.cursor()

query = """SELECT Customer.CustomerId, Customer.FirstName, Customer.LastName, 
Invoice.InvoiceId, InvoiceLine.InvoiceLineId, InvoiceLine.TrackId,
Track.Name, Track.AlbumId, Album.Title
FROM Customer 
INNER JOIN Invoice 
ON
Customer.CustomerId = Invoice.CustomerId
INNER JOIN InvoiceLine
ON InvoiceLine.InvoiceId = Invoice.InvoiceId
INNER JOIN Track
ON Track.TrackId = InvoiceLine.TrackId
INNER JOIN Album
ON Album.AlbumId = Track.AlbumId
ORDER BY Customer.LastName, Customer.FirstName"""

sql_data = pd.read_sql(query, con)
display(sql_data)
```

```python
# Declaring df and columns, removing unneeded columns that were require for SQL join(s), displaying top 5.
df = pd.DataFrame(
    sql_data,
    columns=[
        "CustomerId",
        "FirstName",
        "LastName",
        "InvoiceId",
        "InvoiceLineId",
        "TrackId",
        "Name",
        "AlbumId",
        "Title"
    ])
del df['CustomerId']
del df['InvoiceId']
del df['InvoiceLineId']
del df['TrackId']
del df['AlbumId']
df.head(5)
```
